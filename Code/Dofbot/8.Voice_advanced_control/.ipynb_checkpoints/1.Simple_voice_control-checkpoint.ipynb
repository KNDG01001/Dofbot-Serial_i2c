# 加载机械臂库文件

```python
#!/usr/bin/env python3
#coding=utf-8
import time
from Arm_Lib import Arm_Device

# 获取机械臂的对象
Arm = Arm_Device()
time.sleep(.1)
```

```python
Arm.Arm_ask_speech(9)
```

# 循环检测声音

```python
servo_1 = 90
def main():
    while True:
        global servo_1
        result = Arm.Arm_serial_speech_read(1)
        #print(result)
        if result == 1:
            Arm.Arm_RGB_set(50, 0, 0) #RGB亮红灯
            time.sleep(0.1)  
        elif result == 2:
            Arm.Arm_RGB_set(0, 50, 0) #RGB亮绿灯
            time.sleep(0.1)  
        elif result == 3:
            Arm.Arm_RGB_set(0, 0, 50) #RGB亮蓝灯
            time.sleep(0.1)  
        elif result == 4:              #guandeng 
            Arm.Arm_RGB_set(0, 0, 0)
            time.sleep(0.1) 
            Arm.Arm_Buzzer_On(3)
            time.sleep(.5)
        elif result == 5:                    #xiang shang
            Arm.Arm_serial_servo_write(2, 90, 1000)
            time.sleep(.003)
            Arm.Arm_serial_servo_write(3, 90, 1000)
            time.sleep(.003)
            Arm.Arm_serial_servo_write(4, 90, 1000)
            time.sleep(1)
        elif result == 6:                   #xiang xia 
            Arm.Arm_serial_servo_write(2, 50, 1000)
            time.sleep(.003)
            Arm.Arm_serial_servo_write(3, 50, 1000)
            time.sleep(.003)
            Arm.Arm_serial_servo_write(4, 50, 1000)
            time.sleep(1)
        elif result == 7:                #xiang zuo
            servo_1 += 30
            if servo_1 >= 120:
                servo_1 = 120
            Arm.Arm_serial_servo_write(1, servo_1, 500)
            time.sleep(1)
        elif result == 8:                  #xiang you 
            servo_1 -= 60
            if servo_1 <= 60:
                servo_1 = 60
            Arm.Arm_serial_servo_write(1, servo_1, 500)
            time.sleep(1)
        elif result == 9:              #jia jin
            Arm.Arm_serial_servo_write(6, 180, 500)
            time.sleep(1)
        elif result == 10:             #song kai
            Arm.Arm_serial_servo_write(6, 90, 500)

        time.sleep(0.5)
        #print(" END OF LINE! ")
try :
    main()
except KeyboardInterrupt:
    # 释放Arm对象
    del Arm
    del bus
    print(" Program closed! ")
    pass
```
