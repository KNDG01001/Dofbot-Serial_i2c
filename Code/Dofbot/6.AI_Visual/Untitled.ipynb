```python
#导入相关的模块
import threading
import time
from Arm_Lib import Arm_Device


# 创建机械臂对象
Arm = Arm_Device()
time.sleep(.1)
```

```python
# 定义移动机械臂函数,同时控制1-6号舵机运动，p=[S1,S2,S3,S4,S5,S6]
def arm_move_6(p, s_time = 1500):
    for i in range(6):
        id = i + 1
        Arm.Arm_serial_servo_write(id, p[i], s_time)
        time.sleep(.01)
    time.sleep(s_time/1000)
```

```python
while True:
    arm_move_6(look_at,1500)
    time.sleep(.1)
    arm_move_6(p_Prayer,1500)
    time.sleep(.1)
    arm_move_6(p_Thumb_up,1500)
    time.sleep(.1)
    arm_move_6(p_Heart_single,1500)
    time.sleep(.1)
    arm_move_6(p_Eight,1500)
    time.sleep(.1)
    arm_move_6(p_Congratulation,1500)
    time.sleep(.1)
    arm_move_6(p_Rock,1500)
    time.sleep(.1)
    arm_move_6(p_fist,1500)
    time.sleep(.1)
    arm_move_6(p_horse_1,1500)
    time.sleep(.1)
    arm_move_6(p_horse_2,1500)
    time.sleep(.1)
    arm_move_6(p_horse_3,1500)
    time.sleep(.1)
```
